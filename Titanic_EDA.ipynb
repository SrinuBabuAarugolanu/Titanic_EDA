```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
print(df["Survived"].value_counts())
print(df["Sex"].value_counts())
print(df["Pclass"].value_counts())
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Age'])
plt.title('Age Boxplot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
selected_columns = ['Survived', 'Pclass', 'Age', 'Fare']

sns.pairplot(df[selected_columns].dropna())
plt.show()
```

Summary of Findings

1. More passengers did not survive than survived.
2. Female passengers had a significantly higher survival rate.
3. First-class passengers were more likely to survive.
4. Most passengers were between 20 and 40 years old.
5. Fare distribution is highly skewed with several outliers.
6. Passengers paying higher fares had better survival chances.
7. Cabin contains a large number of missing values.
8. Passenger class and fare are strongly related.
